```python
# import necessary libraries
import pandas as pd
from collections import Counter
from itertools import combinations

# Load the dataset into a pandas dataframe
data = pd.read_excel("./CoffeeShopTransactions.xlsx")
# Preprocess the data
data.drop(['Date', 'Time','Transaction Number'], axis=1, inplace=True)
data = data.iloc[:10, :]
data
```

```python
# Convert data into a list of lists
transactions = []
for i in range(len(data)):
    transaction = []
    for j in range(len(data.columns)):
        if str(data.values[i,j]) != 'nan':
            transaction.append(str(data.values[i,j]))
    transactions.append(transaction)
data = transactions 
```

```python
# Get unique itema from the data
init = []
for i in data:
    for q in i:
        if(q not in init):
            init.append(q)
init = sorted(init)
print(init)
```

```python
# Take minimum support & minimum confidence as a input from user 

support = float (input("Please enter support value: "))   
confidence = float (input("Please enter confidence value: "))
support_Count = int(support*len(data))
support_Count
```

```python
# Applying the apriori algorithm To get Frequent Itemset
c = Counter()
for i in init:
    for d in data:
        if(i in d):
            c[i]+=1
print("C1:")
for i in c:
    print(str([i])+": "+str(c[i]))
print()
l = Counter()
for i in c:
    if(c[i] >= support_Count):
        l[frozenset([i])]+=c[i]
print("L1:")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()
pl = l
pos = 1
for count in range (2,1000):
    nc = set()
    temp = list(l)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            t = temp[i].union(temp[j])
            if(len(t) == count):
                nc.add(temp[i].union(temp[j]))
    nc = list(nc)
    c = Counter()
    for i in nc:
        c[i] = 0
        for q in data:
            temp = set(q)
            if(i.issubset(temp)):
                c[i]+=1
    print("C"+str(count)+":")
    for i in c:
        print(str(list(i))+": "+str(c[i]))
    print()
    l = Counter()
    for i in c:
        if(c[i] >= support_Count):
            l[i]+=c[i]
    print("L"+str(count)+":")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    if(len(l) == 0):
        break
    pl = l
    pos = count
print("Result: ")
print("L"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()
```

```python
# Generating Association Rules
for l in pl:
    c = [frozenset(q) for q in combinations(l,len(l)-1)]
    maxConf = 0
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q)
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa  #confidence
        if(temp > maxConf):
            maxConf = temp
        temp = sab/sb
        if(temp > maxConf):
            maxConf = temp
        rule1 = str(list(a))+" -> "+str(list(b))
        rule2 = str(list(b))+" -> "+str(list(a))
        print(str(list(a))+" -> "+str(list(b))+" = "+str(sab/sa))
        print(str(list(b))+" -> "+str(list(a))+" = "+str(sab/sb))
    curr = 1
    print("choosing:", end=' ')
    choosed = []
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q)
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa
        if(temp >= confidence):
            print(curr, end = ' ')
            choosed.append(rule1)
        curr += 1
        temp = sab/sb
        if( temp >= confidence):
            print(curr, end = ' ')
            choosed.append(rule2)
        curr += 1
    print("\nchoosed rules:")
    for i in choosed:
      print(i)
    print("")
    print()
```
